## Yes. Why?

- It's fast
- Interpreted, but JIT-compiled
- Sophisticated type system, but only if you need it
- Has *Multiple dispatch*, an generalization of OO
- Metaprogramming
- It's fast
- The standard library is written ***in Julia itself***
    - Even primitives like int, rationals, etc
    - Solves the infamous "two language problem" (in many cases)
- It's fast (within 2x-3x C speed)
- Call C functions directly
- MIT licensed

### Familiar syntax

```julia
function fib(n)
    if n <= 1
        return 1
    end
    return fib(n-1) + fib(n-2)
end

for n in [0:15]
    print("$(fib(n)) ")
end
```
